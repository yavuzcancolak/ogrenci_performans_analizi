```python
# Pandas: Tablo (DataFrame) iÅŸlemleri iÃ§in kullanÄ±lÄ±r
import pandas as pd

# Random: Rasgele sayÄ± Ã¼retmek iÃ§in kullanÄ±lÄ±r (Ã¶rn. Ã¶ÄŸrenci notlarÄ±)
import random

# Jupyter'da tabloyu dÃ¼zgÃ¼n gÃ¶stermek iÃ§in display fonksiyonu
from IPython.display import display


# Ã–ÄŸrenci ad-soyad listesi (15 kiÅŸi)
ogrenciler = [
    "AyÅŸe AydÄ±n", "Emre Demir", "Elif YÄ±lmaz", "Kerem Soylu", "Berk YÄ±ldÄ±z",
    "Zeynep KoÃ§", "Ali Ã‡etin", "Merve Aksoy", "Hasan UÃ§ar", "Derya GÃ¼l",
    "Yusuf Toprak", "Buse YalÃ§Ä±n", "Mert KÄ±lÄ±Ã§", "Gizem Akar", "Tolga Can"
]

# Her Ã¶ÄŸrenci iÃ§in veri Ã¼ret
veriler = []
for i, ad_soyad in enumerate(ogrenciler):
    # Cinsiyeti belirle
    cinsiyet = "KadÄ±n" if ad_soyad.split()[0] in ["AyÅŸe", "Elif", "Zeynep", "Merve", "Derya", "Buse", "Gizem"] else "Erkek"

    # 4 haneli ID Ã¼ret (1001'den baÅŸlasÄ±n)
    ogr_id = 1001 + i

    # NotlarÄ± 50-100 arasÄ± random Ã¼ret
    veri = {
        "id": ogr_id,
        "ad_soyad": ad_soyad,
        "cinsiyet": cinsiyet,
        "matematik": random.randint(50, 100),
        "fizik": random.randint(50, 100),
        "kimya": random.randint(50, 100),
        "biyoloji": random.randint(50, 100)
    }
    veriler.append(veri)

# Verileri DataFrame'e aktar
df = pd.DataFrame(veriler)

# Ortalama not sÃ¼tununu ekle
df["ortalama"] = df[["matematik", "fizik", "kimya", "biyoloji"]].mean(axis=1)

# SÄ±nÄ±f listesi gÃ¶rÃ¼nÃ¼mÃ¼ iÃ§in index'i 1'den baÅŸlat
df.index = range(1, len(df)+1)

# CSV olarak dÄ±ÅŸa aktar (4 haneli id ve 1'den baÅŸlayan index ile)
df.to_csv("ogrenciler.csv", index=False)

# ğŸ–¥ SÄ±nÄ±f listesini gÃ¶ster
print("ğŸ“‹ SÄ±nÄ±f Listesi:")
display(df)

# ğŸ” Eksik veri kontrolÃ¼
print("\nğŸ§ª Eksik veri kontrolÃ¼:")
display(df.isnull().sum())

# ğŸ“Š BaÅŸarÄ± sÄ±rasÄ±na gÃ¶re sÄ±ralanmÄ±ÅŸ liste
df_sirali = df.sort_values(by="ortalama", ascending=False)
df_sirali.index = range(1, len(df_sirali)+1)
df_sirali.to_csv("ogrenci_siralamasi.csv", index=False)

print("\nğŸ… BaÅŸarÄ± SÄ±ralamasÄ±:")
display(df_sirali)

```
