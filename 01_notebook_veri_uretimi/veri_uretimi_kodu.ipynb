```python
# Pandas: Tablo (DataFrame) işlemleri için kullanılır
import pandas as pd

# Random: Rasgele sayı üretmek için kullanılır (örn. öğrenci notları)
import random

# Jupyter'da tabloyu düzgün göstermek için display fonksiyonu
from IPython.display import display


# Öğrenci ad-soyad listesi (15 kişi)
ogrenciler = [
    "Ayşe Aydın", "Emre Demir", "Elif Yılmaz", "Kerem Soylu", "Berk Yıldız",
    "Zeynep Koç", "Ali Çetin", "Merve Aksoy", "Hasan Uçar", "Derya Gül",
    "Yusuf Toprak", "Buse Yalçın", "Mert Kılıç", "Gizem Akar", "Tolga Can"
]

# Her öğrenci için veri üret
veriler = []
for i, ad_soyad in enumerate(ogrenciler):
    # Cinsiyeti belirle
    cinsiyet = "Kadın" if ad_soyad.split()[0] in ["Ayşe", "Elif", "Zeynep", "Merve", "Derya", "Buse", "Gizem"] else "Erkek"

    # 4 haneli ID üret (1001'den başlasın)
    ogr_id = 1001 + i

    # Notları 50-100 arası random üret
    veri = {
        "id": ogr_id,
        "ad_soyad": ad_soyad,
        "cinsiyet": cinsiyet,
        "matematik": random.randint(50, 100),
        "fizik": random.randint(50, 100),
        "kimya": random.randint(50, 100),
        "biyoloji": random.randint(50, 100)
    }
    veriler.append(veri)

# Verileri DataFrame'e aktar
df = pd.DataFrame(veriler)

# Ortalama not sütununu ekle
df["ortalama"] = df[["matematik", "fizik", "kimya", "biyoloji"]].mean(axis=1)

# Sınıf listesi görünümü için index'i 1'den başlat
df.index = range(1, len(df)+1)

# CSV olarak dışa aktar (4 haneli id ve 1'den başlayan index ile)
df.to_csv("ogrenciler.csv", index=False)

# 🖥 Sınıf listesini göster
print("📋 Sınıf Listesi:")
display(df)

# 🔎 Eksik veri kontrolü
print("\n🧪 Eksik veri kontrolü:")
display(df.isnull().sum())

# 📊 Başarı sırasına göre sıralanmış liste
df_sirali = df.sort_values(by="ortalama", ascending=False)
df_sirali.index = range(1, len(df_sirali)+1)
df_sirali.to_csv("ogrenci_siralamasi.csv", index=False)

print("\n🏅 Başarı Sıralaması:")
display(df_sirali)

```
